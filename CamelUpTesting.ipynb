```python
player_index = list(Test_0_2.players.keys()).index("Player 1")
print(player_index)

for i in Test_0_2.DO_fields_payoffs.keys():
    field_payoff = Test_0_2.DO_fields_payoffs[i]
    ## delta base is the expected payoff of the players plate minus the base payoff of the plate of the player.
    delta = -Test_0_2.base_DO_hits[player_index,0] +field_payoff[1][player_index] ## create copy of delta base
    ## THIS player additionally gains, if the other players lose hits.
    delta_other_players_hits = np.delete(Test_0_2.base_DO_hits, player_index)-np.delete(field_payoff[1], player_index)

    delta+=np.sum(delta_other_players_hits)

    print(i,Test_0_2.base_DO_hits[player_index,0],-field_payoff[1][player_index],Test_0_2.base_DO_hits[player_index,0] -field_payoff[1][player_index])
    print(np.delete(Test_0_2.base_DO_hits, player_index),np.delete(field_payoff[1], player_index),delta_other_players_hits.sum())
    print(delta, end = ", ")

    #print(Test_0_2.base_payoffs,current_payoffs, delta_payoff_matrix,delta_inventory_payoffs)
    
    current_payoffs = np.matmul(field_payoff[0],np.array([[5,3,2],[1,1,1],[-1,-1,-1]]))
    delta_payoff_matrix = current_payoffs-Test_0_2.base_payoffs
    #print(delta_payoff_matrix)
    delta_inventory_payoffs = np.sum(delta_payoff_matrix*Test_0_2.player_inventory_filter[player_index,:,:])
    delta += delta_inventory_payoffs
    print(delta_inventory_payoffs, delta, end = ", ")

    for player_index_other in range(len(Test_0_2.players)):
        if player_index_other != player_index:
            delta_other_players_bets = np.sum(delta_payoff_matrix*Test_0_2.player_inventory_filter[player_index_other,:,:])
            delta += delta_other_players_bets
            print(delta_other_players_bets, delta, end = ", ")
    print("\n")
```

#### solving problems with camel index finder

```python
rendered_field, inv_player_mask = render_field(demo2.game_field,demo2.players)
camel_row_idx_base = np.zeros(8, dtype=np.int64)
camel_col_idx_base = np.zeros(8, dtype=np.int64)
camels_not_thrown = [1,2,3,4,5,6]
if 6 in camels_not_thrown:  # Extended version (7 camels)
    camel_ids = np.arange(1, 8)
else:
    camel_ids = np.arange(1, 6)
for i in camel_ids:
    pos = np.argwhere(rendered_field == i)
    camel_row_idx_base[i] = pos[0][0]
    camel_col_idx_base[i] = pos[0][1]
```
