## Starting field scenarios
The starting field has a very limited setup. It is worth looking at the most common situations here,
<br> as we will always play one of these setups at the start of the game, meaning we can already get a 
<br> heads start, if we know in descending order the best moves available here. Consider this the same way 
<br> as openings in chess, only that here there are probably dominant strategies. Spoiler alert: throwing
<br> the dice is not one of them.
<br>
<br>Fundamentally, there are 5**3=243 paths for the starting position of camels.
<br>The following dictionary shows the results for starting position and the respective paths

```python
start_scenarios = {
    (5,0,0):1, (0,5,0):1, (0,0,5):1, # (5 chose 0 for all of them)
    (4,1,0):5, (4,0,1):5, (1,4,0):5, (0,4,1):5, (1,0,4):5, (0,1,4):5, # (5 chose 1 for all of them)
    (3,1,1):20, (1,3,1):20, (1,1,3):20, # 5!/(1*1*3!)
    (3,2,0):10, (3,0,2):10, (2,3,0):10, (0,3,2):10, (2,0,3):10, (0,3,2):10, # (5!/(2!*3!))
    (2,2,1):30, (2,1,2):30, (1,2,2):30} # 5!/(2!*2!*1!)
```

Since in the beginning of the game, it is very unlikely for the game to end round 1, even without 
<br> player involvement through desert and oasis field, [5,0,0], [0,5,0], and [0,0,5] are equivalent. 
<br>The same goes for [3,2,0] and [0,3,2] for example.
<br>
<br> The most important things to factor in at the beginning of the game are desert/oasis placement and 
<br> early betting patterns. These may persist for similar scenarios throughout the game, probably at 
<br> least for round 2.
<br>
### Scenario [2,2,1]
#### [DESERT 4], [DESERT 6]
<br> Lets look at some concrete examples. Most commonly for the early game, will be cases like [2,2,1], [2,1,2], 
<br> and [1,2,2], respectively. Lets walk through some move patterns pre first roll.

```python
## define our base players
Players = {
    "Player 1":player("Player 1"),"Player 2":player("Player 2"),
    "Player 3":player("Player 3"),"Player 4":player("Player 4")}
Field_2_2_1 = Field([
    ['YELLOW', 'GREEN'], ['PURPLE', 'BLUE'], ['ORANGE'], [], [], [], [], [], [], 
    [], [], [], [], [],[], ["BLACK", "WHITE"], [], [], []],copy.deepcopy(Players))

## Init function with pre-defined players and game field
Scenario_2_2_1_01 = CamelUp(4,field = Field_2_2_1,black_white=True,user_guide=True)
Scenario_2_2_1_01.game("Player 1")
```

At the first move, the top 2 camels in terms of EV are Blue with 1.33 and Green with 1.07, closely followed by Yellow with 1.00. 
<br> There is also the question for the player wether to place the desert/oasis early, especially because the EV of field 4 is 1.32.
<br> The obvious pick here is [DESERT 4], as it has high EV and the other players cannot mess with it. This may be followed commonly 
<br> by [DESERT 6], which has a low EV now, but later might have increased value through Control, meaning the OPTION to control 
<br> (expected) payoffs of all players later, should the camels move around the field. Control is probably worth less on field 4 
<br> than it is on 5 or 6, as it is quite likely that at least one Camel will jump it. Field 5 has another unseen and hard to price 
<br> advantage of blocking both 4 and 6 for placement for other players.
<br>
<br> Lets go with [DESERT 4] this time, as I think for most players this is the most likely move in this case, because it is such an 
<br> obvious maximum EV pick and it also directly shows us the impact on the payoffs. Thereafter I will select [DESERT 6] for the 
<br> second move.
<br>
<br> [DESERT 4] decreases the value of the maximum EV bet [BET5 2.1] from 1.33 to 1.15 and raises [BET5 3.1] from .77 to 1. It also
<br> lowers [BET5 3.1 under 1]. [DESERT 6] reverses the effect on C2.1 slightly, but further decreases expected payoffs of C3.1 and C1.1.
<br>
#### [DESERT 5]
<br> Another possibility in this case is for a Player to play [DESERT 5]. Lets look at that scenario:

```python
Players = {
    "Player 1":player("Player 1"),"Player 2":player("Player 2"),
    "Player 3":player("Player 3"),"Player 4":player("Player 4")}
Field_2_2_1 = Field([
    ['YELLOW', 'GREEN'], ['PURPLE', 'BLUE'], ['ORANGE'], [], [], [], [], [], [], 
    [], [], [], [], [],[], ["BLACK", "WHITE"], [], [], []],copy.deepcopy(Players))
## Init function with pre-defined players and game field
Scenario_2_2_1_02 = CamelUp(4,field = Field_2_2_1,black_white=True,user_guide=True)
Scenario_2_2_1_02.game("Player 1")
```

[DESERT 5] changes the payoffs quite a bit. It lowers the expected payoffs of C3.1 and C2.1, and puts C1.1 on top at 1.17 EV.
<br> [DESERT 5] a lower EV than [DESERT 4], but it increases uncertainty of payoffs in the field. It also lowers VOI and thus 
<br> increases other players willingness to role the dice.

#### [BET5 2.1] [BET5 1.1], [OASIS 4], [DESERT 6]

### Scenario [2,1,2]
#### [BET5 3.1] [BET3 3.1] [OASIS 4] [DESERT 6]

```python
Players = {
    "Player 1":player("Player 1"),"Player 2":player("Player 2"),
    "Player 3":player("Player 3"),"Player 4":player("Player 4")}
Field_2_2_1 = Field([
    ['YELLOW', 'GREEN'], ['PURPLE'], ['ORANGE', 'BLUE'], [], [], [], [], [], [], 
    [], [], [], [], [],[], [], [], [], ["BLACK", "WHITE"]],copy.deepcopy(Players))
## Init function with pre-defined players and game field
Scenario_2_1_2_01 = CamelUp(4,field = Field_2_2_1,black_white=True,user_guide=True)
Scenario_2_1_2_01.game("Player 1")
```

C3.1 has a staggering 2.45 EV starting in [2,1,2], meaning [BET5 3.1] is an obvious pick one, as is [BET3 3.1] a very good second pick
<br> with an EV of 1.46 at this point. EV maximising players play [OASIS 4] and [DESERT 6], but there is also an option to play [DESERT 5]:
#### [BET5 3.1] [BET3 3.1] [DESERT 5] 

```python
Players = {
    "Player 1":player("Player 1"),"Player 2":player("Player 2"),
    "Player 3":player("Player 3"),"Player 4":player("Player 4")}
Field_2_2_1 = Field([
    ['YELLOW', 'GREEN'], ['PURPLE'], ['ORANGE', 'BLUE'], [], [], [], [], [], [], 
    [], [], [], [], [],[], [], [], [], ["BLACK", "WHITE"]],copy.deepcopy(Players))
## Init function with pre-defined players and game field
Scenario_2_1_2_02 = CamelUp(4,field = Field_2_2_1,black_white=True,user_guide=True)
Scenario_2_1_2_02.game("Player 1")
```

It is better for Player 3 to leave the space and put [OASIS 4], as this minimizes the EV of C3.1 and puts him only .6 behind Player 1
<br> A rational Player 4 will collaborate and play [DESERT 6]!
