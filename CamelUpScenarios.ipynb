## Starting field scenarios
The starting field has a very limited setup. It is worth looking at the most common situations here,
<br> as we will always play one of these setups at the start of the game, meaning we can already get a 
<br> heads start, if we know in descending order the best moves available here. Consider this the same way 
<br> as openings in chess, only that here there are probably dominant strategies. Spoiler alert: throwing
<br> the dice is not one of them.
<br>
<br>Fundamentally, there are 5**3=243 paths for the starting position of camels.
<br>The following dictionary shows the results for starting position and the respective paths

```python
start_scenarios = {
    [5,0,0]:1, [0,5,0]:1, [0,0,5]:1, # (5 chose 0 for all of them)
    [4,1,0]:5, [4,0,1]:5, [1,4,0]:5, [0,4,1]:5, [1,0,4]:5, [0,1,4]:5, # (5 chose 1 for all of them)
    [3,1,1]:20, [1,3,1]:20, [1,1,3]:20, # 5!/(1*1*3!)
    [3,2,0]:10, [3,0,2]:10, [2,3,0]:10, [0,3,2]:10, [2,0,3]:10, [0,3,2]:10, # (5!/(2!*3!))
    [2,2,1]:30, [2,1,2]:30, [1,2,2]:30} # 5!/(2!*2!*1!)
```

Since in the beginning of the game, it is very unlikely for the game to end round 1, even without 
<br> player involvement through desert and oasis field, [5,0,0], [0,5,0], and [0,0,5] are equivalent. 
<br>The same goes for [3,2,0] and [0,3,2] for example.
<br>
<br> The most important things to factor in at the beginning of the game are desert/oasis placement and 
<br> early betting patterns. These may persist for similar scenarios throughout the game, probably at 
<br> least for round 2.
<br>
<br> Lets look at some concrete examples. Most commonly for midgame, and also really common early game,
<br> will be cases like [3,0,2] or [2,0,3], as midgame they will arise often from desert/oasis placement.
